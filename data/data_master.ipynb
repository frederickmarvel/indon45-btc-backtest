```python
import pandas as pd

df_price = pd.read_csv('indon/indonprice.csv')
df_yield = pd.read_csv('indon/indonyield.csv')

df_price['price_date'] = pd.to_datetime(df_price['price_date'])
df_yield['price_date'] = pd.to_datetime(df_yield['price_date'])

merged_df = pd.merge(df_price, df_yield, on='price_date', suffixes=('_price', '_yield'))

print(merged_df.head())

merged_df.to_csv('indon_merged_data.csv', index=False)
```

```python
import pandas as pd

# 1. Load your existing merged data
df_merged = pd.read_csv('indon_merged_data.csv')
df_merged['price_date'] = pd.to_datetime(df_merged['price_date'])

# 2. Load BTC data
# We tell pandas to treat '.' as thousands and ',' as decimal immediately
btc = pd.read_csv('btcusd_2015_2026.csv', thousands='.', decimal=',')

# 3. Clean up BTC columns
# Keep only what you need and rename them to match your main dataframe
btc = btc[['Tanggal', 'Terakhir']].copy()
btc.columns = ['price_date', 'btc_close']

# 4. Standardize dates
# 'dayfirst=True' is essential for DD/MM/YYYY format
btc['price_date'] = pd.to_datetime(btc['price_date'], dayfirst=True)

# 5. Final Merge
final_df = pd.merge(df_merged, btc, on='price_date', how='left')

# Show result
print(final_df.head())
final_df.to_csv('final_dataset.csv', index=False)
```

```python
df = pd.read_csv('final_dataset.csv')
df_cleaned = df.drop_duplicates()
df_cleaned.to_csv('final_dataset_cleaned.csv', index=False)
```
